**Using Pandas**

We start out by practicing how to handle data given to us. The rows in any dataset are called tuples or instances and the columns are the features with one of the columns being the label (there may also not be a label if it is an Unsupervised learning task). 

Storing this data as a list of lists might be cumbersome and difficult depending on the complexity and size of the data.

Pandas offers a data structure called a DataFrame, which makes it easy to handle data and visualise the intricacies of it.

**Using Matplotlib**

It is important to visualise the data we have so we can get useful inferences from it, such as by plotting one feature against another. If we can guess beforehand which features are more likely to decide the value of the target variable, we'll be able to create a better model accordingly.

```matplotlib.pyplot``` is the library for creating 2D graphs in Python. It mimics the functions used in Matlab, which is a much more comprehensive tool for a wide range of experimental purposes.

````python
import matplotlib.pyplot as plt

# Only for jupyter-notebooks; don't bother about this line elsewhere.
%matplotlib inline
````

We'll try to plot the features from the Iris dataset.

````python
sl = df['sepal length (cm)']
````

````python
# Create a plt.figure object
fig = plt.figure('Sepal length vs y')

ax0 = fig.add_subplot(111) # This means there's only 1 plot in the whole figure. 211 would mean 2 plots
# and the top plot is being added. 212 means 2 plots and bottom plot is being added.
ax0.scatter(sl, y, c = 'b', marker='o', label='Sepal length')
ax0.set_xlabel('sepal length (cm)')
ax0.set_ylabel('flower')
ax0.legend()

plt.show()
````

````python
# Plot petal width vs y with blue o's and petal length vs y with red x's
fig = plt.figure('Petal width vs y,Petal length vs y')

### CODE HERE ###
ax0 = fig.add_subplot(111)
pw=df['petal width (cm)']
pl=df['petal length (cm)']
ax0.scatter(pw, y, c = 'b', marker='o', label='Sepal length')
ax0.scatter(pl,y,c='r',marker='x',label='Petal length')
ax0.set_xlabel('length/width')
ax0.set_ylabel('flower')
ax0.legend()

plt.show()
````

````python
# Plot sepal length vs petal length for each instance
fig = plt.figure('Sepal length vs petal length')

### CODE HERE ###
ax0=fig.add_subplot(111)

ax0.scatter(sl,pl,c='m',marker='o',label='Sepal length')
ax0.set_xlabel("sepal length")
ax0.set_ylabel("petal length")
ax0.legend()

plt.show()
````

**Using Scikit-Learn**

Scikit-Learn offers a variety of classifiers for learning the data and creating a model for it. This task is a classification task and hence a regression method such as Linear Regression cannot be used for it. We'll use Logistic Regression to create a classifier for this data. 

Before using a model on the data, we will need some data to test the model on. For this, we'll keep some data aside called the 'test set'.

Also, we can have some problems in learning the data if the data is not randomly shuffled and chosen first. (More on this later)

````python
# Join target variable to dataframe
### CODE HERE ###
from sklearn.datasets import load_iris
import pandas as pd
data = load_iris()
X = data['data']
header=data['feature_names']
df=pd.DataFrame(X,columns=header)
header=data['target']
df2=pd.DataFrame(header,columns=['target'])
df['target']=df2
# Randomly shuffle dataframe
### CODE HERE ###
````

````python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

# We divide df into training and testing sets such that the training set is 80% of df and test set is 20%
train, test = train_test_split(df, test_size=0.2)

y_train = train['target']
y_test = test['target']
X_train = train[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']]
X_test = test[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']]

# Create a classifier
clf = LogisticRegression()
# Train it on the training set
clf.fit(X_train, y_train)
````

````python
# Calculate the predictions on the test set
pred = clf.predict(X_test)

pred
````

````python
from sklearn.metrics import accuracy_score

# Find the accuracy on the test set
accuracy_score(pred, y_test)
````

Since the accuracy is high, we can assume this is a good model we have created. However, there are a lot of factors that decide how relevant the accuracy is to the actual performance of the model. (Again, more on this later!)
